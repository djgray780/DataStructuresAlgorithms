```python
class Empty(Exception):
    """Error when trying to access element from an empty container"""
    pass

class Stack:
    """ LIFO Stack implementation using Adapter pattern for lists """
    def __init__(self):
        """ Define an empty stack """
        self._data = []

    def __len__(self):
        """ Return the number of elements in Stack"""
        return len(self._data)

    def push(self, element):
        """ Add element to top of the Stack"""
        self._data.append(element)

    def pop(self):
        """ Remove & return the top element from the stack, raise Empty exception if Stack is empty """
        if self.is_empty():
            raise Empty('Stack is empty')
        return self._data.pop()

    def is_empty(self):
        """ Return True if the Stack is empty """
        if len(self.data) == 0: # Can make this a single line, but this is more explicit imo
            return True
        return False

    def top(self):
        """ Return, but do not remove, the element at the top of stack. Raise Empty if stack empty"""
        if self.is_empty():
            raise Empty('Stack is empty')
        return self._data[-1]
```

```python
class Empty(Exception):
    """Error when trying to access element from an empty container"""
    pass

class StackArray():
    # Constructor
    def __init__(self):
        """Define the empty container"""
        self._container = []

    # Push onto the stack
    def push(self, e):
        """Add element to top of the stack"""
        self._container.append(e)

    # Pop off of the stack
    def pop(self):
        """Remove top element from the stack"""
        if self.is_empty():
            raise Empty('Stack is empty')
        return self._container.pop()

    # Peak at the top of the stack
    def peak(self):
        if self.is_empty():
            raise Empty('Stack is empty')
        return self._container[-1]

    # Check if stack is empty
    def is_empty(self):
        """Returns True if empty, False otherwise"""
        return len(self._container) == 0

    # Return length of the stack
    def __len__(self):
        """Return length of the Stack"""
        return len(self._container)
```
